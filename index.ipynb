```python
from imblearn.under_sampling import RandomUnderSampler
import imblearn.keras as keras
from sklearn.model_selection import *
from sklearn.linear_model import *
from sklearn.preprocessing import *
import pandas as pd
sampler=RandomUnderSampler()
model=PolynomialFeatures(degree=1,include_bias=True)
dataset=pd.read_csv("biasing.csv")
dataset
```

```python
# 
accuracies=[]
i=0.1
j=0.01
while (i<0.9):
    x1,x2,y1,y2=train_test_split(dataset[["Feature1","Feature2","Feature3","Feature4","Feature5"]],dataset["Output"],test_size=i)
    model2=LinearRegression()
    model2.fit(x1,y1)
    model2.predict(x2)
    score=model2.score(x2,y2)
    accuracies.append((i,score))
    i+=j
```

```python
data=pd.DataFrame(model.fit_transform(dataset[["Feature1","Feature2","Feature3","Feature4","Feature5"]],dataset["Output"]),columns=dataset.columns)
data
x1,x2,y1,y2=train_test_split(data[["Feature1","Feature2","Feature3","Feature4","Feature5"]],data["Output"],test_size=0.2)
a=list(x[1] for x in accuracies)
print(accuracies[a.index(max(a))])
```
