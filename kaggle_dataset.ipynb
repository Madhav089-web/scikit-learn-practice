```python
import pandas as pd
unclean_data=pd.read_csv("unclean_dataset.csv")
one=int(unclean_data["feature_1"].value_counts()[0])
two=int(unclean_data["feature_1"].value_counts()[2])
three=int(unclean_data["feature_1"].value_counts()[3])
five=int(unclean_data["feature_1"].value_counts()[4])
one,two,three,five
# unclean_data["feature_1"].value_counts()[4]
```

```python
def editMissing(x):
    y=str(x["feature_1"])
    avg=(1*one+2*two+3*three+5*five)/(one+two+three+five)
    if(y=='missing'):
        return avg
    elif(y=='nan'):
        return avg
    else:
        a=eval(y)
        return float(a)
    # print(y)
unclean_data["feature_1"]=unclean_data.apply(editMissing,axis=1)
unclean_data
```

```python
fifty=int(unclean_data["feature_2"].value_counts()[0])
ten=int(unclean_data["feature_2"].value_counts()[1])
twenty=int(unclean_data["feature_2"].value_counts()[2])
thirty=int(unclean_data["feature_2"].value_counts()[4])
fifty,ten,twenty,thirty
unclean_data["feature_2"].value_counts()
```

```python
def cleanUnknown(x):
    y=x
    avg=(50*fifty+20*twenty+10*ten+30*thirty)/(fifty+twenty+ten+thirty)
    if(y=="unknown"):
        return avg
    else:
        return float(y)
unclean_data['feature_2']=unclean_data['feature_2'].apply(cleanUnknown)
```

```python
relLifeData=pd.read_csv("real_life_dataset.csv")
relLifeData
```

```python
from sklearn.preprocessing import *
from sklearn.model_selection import *
from sklearn.linear_model import *
data=relLifeData[["Age","Annual_Income","Purchase_Amount","Purchased"]]
data
```

```python
# import matplotlib.pyplot as plt
# plt.plot(data["Annual_Income"],data["Purchase_Amount"])
# plt.show()
data["Annual_Income"]=data["Annual_Income"].apply(lambda x:float(x))
data["Annual_Income"]
```

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
from sklearn.datasets import *
model=LogisticRegression()
x,y=make_classification(n_samples=1000,n_features=2,n_informative=2,random_state=42,n_redundant=0,n_repeated=0)
model.fit(data["Annual_Income"],data["Purchase_Amount"])
plot_decision_regions(x, y, clf=model, legend=2)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Decision Region using Logistic Regression')
plt.show()
```
